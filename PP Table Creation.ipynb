```python
import os
import pandas as pd
import json
```

```python
# Aggregated Transaction
path1 = "C:/Users/admin/Desktop/Guvi-Data Science/Streamlit/pulse/data/aggregated/transaction/country/india/state/"
agg_state_list = os.listdir(path1)

data_dict3 = {'State': [], 'Year': [], 'Quater': [], 'Transaction_type': [], 'Transaction_count': [], 'Transaction_amount': []}

transactions = [d for d in agg_state_list if os.path.isdir(os.path.join(path1, d))]

for state in transactions:
    state_path = os.path.join(path1, state)
    years = [os.path.join(state_path, y) for y in os.listdir(state_path) if os.path.isdir(os.path.join(state_path, y))]

    for year_path in years:
        files = [os.path.join(year_path, f) for f in os.listdir(year_path) if os.path.isfile(os.path.join(year_path, f))]

        for file_path in files:
            with open(file_path, 'r') as data_file:
                A= json.load(data_file)
                for item in A['data']['transactionData']: 
                    trans_type = item['name']
                    trans_count = item['paymentInstruments'][0]['count']
                    trans_amount = item['paymentInstruments'][0]['amount']
                    data_dict3['Transaction_type'].append(trans_type)
                    data_dict3['Transaction_count'].append(trans_count)
                    data_dict3['Transaction_amount'].append(trans_amount)
                    data_dict3['State'].append(state)
                    data_dict3['Year'].append(os.path.basename(year_path))
                    data_dict3['Quater'].append(int(os.path.splitext(os.path.basename(file_path))[0]))

# Successfully created a DataFrame
agg_transaction_df = pd.DataFrame(data_dict3)
agg_transaction_df
```

```python
# Map Transaction 
path3 = "C:/Users/admin/Desktop/Guvi-Data Science/Streamlit/pulse/data/map/transaction/hover/country/india/state/"
map_hover_state_list = os.listdir(path3)

data_dict_map_hover_transaction = {'State': [], 'Year': [], 'Quater': [], 'Transaction_type': [], 'Transaction_count': [], 'Transaction_amount': []}

map_hover_transactions = [d for d in map_hover_state_list if os.path.isdir(os.path.join(path3, d))]

for state in map_hover_transactions:
    state_path = os.path.join(path3, state)
    years = [os.path.join(state_path, y) for y in os.listdir(state_path) if os.path.isdir(os.path.join(state_path, y))]

    for year_path in years:
        files = [os.path.join(year_path, f) for f in os.listdir(year_path) if os.path.isfile(os.path.join(year_path, f))]

        for file_path in files:
            with open(file_path, 'r') as data_file:
                map_hover_data = json.load(data_file)
                for item in map_hover_data['data']['hoverDataList']:
                    for metric in item['metric']:
                        trans_type = f"{item['name']} - {metric['type']}"
                        trans_count = metric['count']
                        trans_amount = metric['amount']
                        data_dict_map_hover_transaction['Transaction_type'].append(trans_type)
                        data_dict_map_hover_transaction['Transaction_count'].append(trans_count)
                        data_dict_map_hover_transaction['Transaction_amount'].append(trans_amount)
                        data_dict_map_hover_transaction['State'].append(state)
                        data_dict_map_hover_transaction['Year'].append(os.path.basename(year_path))
                        data_dict_map_hover_transaction['Quater'].append(int(os.path.splitext(os.path.basename(file_path))[0]))

# Successfully created a DataFrame
map_hover_transaction_df = pd.DataFrame(data_dict_map_hover_transaction)
map_hover_transaction_df
```
